```python
import pandas as pd
```

```python
Org = pd.read_csv('TestData.csv') #載入資料
```

```python
Org.head()
```

```python
from collections import Counter
```

```python
Names = Counter(Org['legislatorName']) #列出有發言的立委名單，以及發言次數
Names
```

```python
from operator import itemgetter
sorted(Names.items()) #先依照姓名排序，用來整併名稱重複的立委
```

```python
Names['伍麗華Saidhai Tahovecahe'] = Names['伍麗華Saidhai Tahovecahe'] + Names['伍麗華Saidhai‧Tahovecahe'] 
```

```python
Names['伍麗華Saidhai Tahovecahe']
```

```python
del Names['伍麗華Saidhai‧Tahovecahe']
```

```python
SortNames = sorted(Names.items(), key=lambda pair: pair[1], reverse=True) #依照發言次數由多至少排序
SortNames
```

```python
maxName = SortNames[0] #分析最多發言的立委 - 陳椒華
maxName[0]
```

```python
#找出陳椒華都在哪些會議上發言
Chen = Org.loc[Org['legislatorName'] == maxName[0]]
ChenTalkNumber = sorted(Counter(Chen['meetingName']).items(), key=lambda pair: pair[1], reverse=True)
ChenTalkNumber[0][0]
```

```python
#找出發言最多的會議，是在談甚麼議題
ChenMeeting = Org.loc[ (Org['legislatorName'] == maxName[0]) & (Org['meetingName'] == ChenTalkNumber[0][0]) ]
meetingContent = list(Counter(ChenMeeting['meetingContent']).keys())[0]
meetingContent
```

```python
#此議題還有哪些立委有發言關注
meetings = Org.loc[Org['meetingContent'] == meetingContent]
sorted(Counter(meetings['legislatorName']).items(), key=lambda pair: pair[1], reverse=True)
```
