輸入立委臉書粉專內容及選舉政見資料

```python
import pandas as pd
fb = pd.read_csv('Xie.csv')
len(fb)
```

```python
fb.head(1)
```

```python
fb[fb.created_time_taipei == max(fb.created_time_taipei)]
```

```python
fb[fb.created_time_taipei == min(fb.created_time_taipei)]
```

```python
fb['new_date'] = pd.to_datetime(fb['created_time_taipei']).dt.date
fb['month_year'] = pd.to_datetime(fb['new_date']).dt.to_period('M')
fb.head(1)
```

```python
results = fb.groupby('month_year').sum()
results
```

```python
results.index.name = 'newhead'
results.reset_index(inplace=True)
results
```

```python
results['newhead'].map(str)
```

```python
import matplotlib.pyplot as plt

labels = results['newhead'].map(str)
width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.bar(labels, results['like_count'], width, label='like_count')
ax.bar(labels, results['share_count'], width, label='share_count', bottom=results['like_count'])
ax.bar(labels, results['comment_count'], width, label='comment_count', bottom=results['share_count']+results['like_count'])

ax.legend()

plt.show()
```

討論此聲量變化的可能原因。
今年四月疫情爆發前，聲量從低往上走，想知道是為什麼？

```python
April = fb[fb.month_year == "2021-04"]
```

```python
resultsApril = April.groupby('new_date').sum()
resultsApril
```

```python
resultsApril.index.name = 'newhead'
resultsApril.reset_index(inplace=True)
resultsApril
```

```python
df_add = pd.DataFrame({'newhead':['2021-04-05'],'like_count':[0],'comment_count':[0],'share_count':[0]})
```

```python
def insert(df, i, df_add):
    df1 = df.iloc[:i, :]
    df2 = df.iloc[i:, :]
    df_new = pd.concat([df1, df_add, df2], ignore_index = True)
    return df_new
```

```python
df_new = insert(resultsApril, 4, df_add)
df_new
```

```python
labels = [str(i) for i in range(1,31)]
labels
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.plot(labels, df_new['share_count'], color='red')
ax.plot(labels, df_new['comment_count'], color='blue')

plt.show()
```

```python
# 用【紓困】來搜尋謝衣鳳臉書相關發文
bool = fb['message'].str.contains('紓困')
bool.value_counts()
```

```python
type(bool)
```

```python
idx = bool.array
idx
```

```python
import numpy as np
A_true = np.where( idx )
A_true
```

```python
covid = fb.loc[A_true]
covid
```

```python
labels = covid['new_date'].map(str)
width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.bar(labels, covid['like_count'], width, label='like_count')
ax.bar(labels, covid['share_count'], width, label='share_count', bottom=covid['like_count'])
ax.bar(labels, covid['comment_count'], width, label='comment_count', bottom=covid['share_count']+covid['like_count'])

ax.legend()

plt.show()
```
