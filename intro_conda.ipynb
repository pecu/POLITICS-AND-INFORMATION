## 開發環境控制

## cmd 操作

+ 看目前的環境有哪些</br>
conda env list

+ 創環境；–n後面填自己要的名字(ex:my_env)；最後面可以指定版本 </br>
conda create –name my_env python=3.6

+ 進入環境
    + windows </br>
     activate my_env
    + macOS/linux </br>
    source activate my_env

+ 離開環境</br>
    + windows </br>
    conda deactivate
    + macOS/linux </br>
    source deactivate

+ 刪掉環境裡面的某個套件</br>
conda remove --name my_env numpy

+ 刪掉整個環境</br>
conda env remove --name my_env
