## LDA-TM : 
### Extract Topic Model with LdaModel from KCC News and CIS test data

```python
# -*- coding: utf-8 -*-
"""
Input: "KCC Data/CkipNewsCisTest-KccDict2020.txt"
Output: "KCC Data/NewsCisTest-KccDict2020-LDA-TopicNum10.txt"

@author: Johnson
"""
#一開始先把套件讀進來
from gensim import corpora, models, utils
from gensim.models import LdaModel, CoherenceModel
```

###  Read data from txt file, and tokenize words, clean-up text

+ 匯入字典
+ 「CkipNewsCisTest-KccDict2020.txt」
    + 先讀進來一小段文字(當年馬英九到衛武營的報導)
    + 檔案已經斷好詞也清好停用字了

```python
fList = []
#先把檔案讀進來
with open("../KCC Data/News-CIS Data/CkipNewsCisTest-KccDict2020.txt","r", encoding='utf-8-sig') as files:
     for file in files:
        #simple_preprocess: tokenize,去除標點 - gensim.utils.simple_preprocess(doc, deacc=False, min_len=2, max_len=15) 
        fList.append(list(utils.simple_preprocess(file, deacc=True, min_len=2,)))
print("fList[:1] = ", fList[:1])
print("------------------------------")
```

### Creat dictionary for corpus
+ 設置字典

```python
dict = corpora.Dictionary(fList)
print("dict = ", dict)
print("------------------------------")
```

### Creat corpus need for Topic Modeling
+ 設定好字典之後，我們把要分析的語料庫(corpus)匯進來</br>
我們看看每個字出現的頻率有幾次

```python
# creat mapping matrix of [word_id, word_frequency] for each word segment
corpus = [dict.doc2bow(text) for text in fList]   

#上面這邊印出每個字的詞頻
#(a,b) a是編號第幾個字 / b是那個字出現了幾次
print("corpus[:1] = ", corpus[:1])   
print("------------------------------")

# 下面是哪個字具體出現了幾次
print("print lists of [word:frequency]")
[[print((dict[id], freq)) for id,freq in cp] for cp in corpus[:1]]
print("================end of the ouput================")
```

### Building the Topic Model from LdaModel

+ Latent Dirichlet Allocation（LDA）
屬於非監督式模型，可以將字詞依據相似程度，分成不同的主題主題。</br></br>

+ LDA怎麼看

    + 以第一個主題為例：
    strTM =  00.022*"台灣" + 0.022*"日本" + 0.016*"周玲妏" + 0.016*"學校" + 0.010*"空間" + 0.010*"熊本" + 0.010*"時間" + 0.010*"整併" + 0.010*"市長" + 0.008*"團隊" + 0.008*"市府" + 0.008*"網友" + 0.008*"特別" + 0.008*"港口" + 0.006*"市民" + 0.006*"同事" + 0.006*"支持" + 0.006*"政府" + 0.006*"議員" + 0.006*"縣市合併"

    strTM列出的就是主題的代表字，字詞前面的數字代表的就是這個字在這個主題裡面的重要程度。如「台灣」這個字在主題一的權重就是「00.022」。

+ LDA 模型的衡量方式
    + 通常仰賴人工判定分群效果(句否有聚集特定社會事件的相關詞彙)

    + Perplexity：可評估模型選了多少主題</br>
    通常較小的Perplexity代表較好的模型，但仍要看容不容易解讀而定。
    
    + Coherence Score: 一致性</br>
    越高的一致性，代表模型內的東西越相似


```python
# 設定模型分類的參數
lda_model = models.ldamodel.LdaModel(corpus=corpus, id2word= dict, num_topics=10, passes=20, alpha='auto', chunksize=50, per_word_topics=True)
output = open("../KCC Data/NewsCisTest-KccDict2020-LDA-TopicNum10.txt", 'w', encoding='utf-8-sig')

#把分好的主題關鍵字印出來 #印10個主題，一個主題20個字
lstTM = lda_model.print_topics(num_topics=10, num_words=20)
for tupleTM in lstTM:
        str1 = "".join(str(x) for x in tupleTM)
        #print("type(str1) = ", type(str1))
        str1 += "\n"
        print("strTM = ", str1) #把主題內的代表字印出來
        output.write(str1)
        
output.close()
 
print("------------------------------")
print("Perplexity = ", lda_model.log_perplexity(corpus))

print("------------------------------")
cm = CoherenceModel(model=lda_model, texts=list(fList), dictionary=dict, coherence= 'c_v' )
coher_lda = cm.get_coherence()
print("\n Coherence Score: ", coher_lda) #印出聚合分數，越高越好
```

### Visualize the topics-keywords
這邊的圖形來自於上面的分群結果「lda_model」，</br>
越大的氣泡代表主題越普遍，滑鼠移到泡泡上面的時候，變成紅色的字就是組成主題的字。</br>
通常好的分群模型，氣泡之間會分的比較開。

```python
import pyLDAvis
import pyLDAvis.gensim
import matplotlib.pyplot as plt
%matplotlib inline

pyLDAvis.enable_notebook
vis = pyLDAvis.gensim.prepare(lda_model, corpus, dict, doc_topic_dist=None)
"""
參數說明：
準備（topic_model，語料庫，字典，doc_topic_dist =無）
topic_model：訓練得到的gensim LdaModel對象，不支持其他gensim模型類型
語料庫：以詞袋形式表示的語料（用於訓練主題模型的語料）
字典：用於造詞袋模型的字典，即gensim字典對象
doc_topic_dist：可選參數，用於LDA模型的文檔主題分佈，替換取值無，當需要多次調用prepare函數時，可以替換該參數
"""
pyLDAvis.display(vis)
```
